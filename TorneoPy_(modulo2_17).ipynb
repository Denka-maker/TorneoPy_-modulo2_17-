```python
#Listas
jugadores = ["Player1", "Player2", "Player3"]
puntajes  = [80, 40, 95]

#Jugador con mayor puntaje
max_score = max(puntajes)
idx_max = puntajes.index(max_score)
jugador_top = jugadores[idx_max]

#Diccionario jugador -> puntaje
jugador_dict = dict(zip(jugadores, puntajes))

#Top 3 ordenado de mayor a menor
#Emparejar, ordenar y asignar ranking
pares = list(zip(jugadores, puntajes))
pares_ordenados = sorted(pares, key=lambda x: x[1], reverse=True)  # de mayor a menor
top3 = pares_ordenados[:3]

# Mostrar resultados
print(f"Jugador con mayor puntaje: {jugador_top} ({max_score} puntos)\n")
print("Diccionario jugador -> puntaje:")
print(jugador_dict, "\n")
print("Top 3 (ordenado de mayor a menor):")
for r, (j, p) in enumerate(top3, start=1):
    print(f"{r}. {j} - {p} puntos")
```

```python
import random
import pandas as pd
from pathlib import Path

try:
    from caas_jupyter_tools import display_dataframe_to_user
    has_display_tool = True
except Exception:
    has_display_tool = False

#Definir la función reutilizable
def calcular_puntaje(kills, asistencias):

    return kills * 10 + asistencias * 5

#Definir jugadores y estadísticas iniciales
jugadores = [
    "Aldo", "Beatriz", "Carlos", "Diana", "Esteban", "Fernanda", "Gabriel", "Helena"
]

estadisticas = {p: {"kills": 0, "asistencias": 0, "puntos_totales": 0, "rondas_pasadas": 0} for p in jugadores}

NUM_RONDAS = 5
RANDOM_SEED = 42

random.seed(RANDOM_SEED)

for ronda in range(1, NUM_RONDAS + 1):
    for p in jugadores:
        #estadísticas por ronda
        kills_r = random.randint(0, 10)
        asist_r = random.randint(0, 8)
        puntos_r = calcular_puntaje(kills_r, asist_r)

        estadisticas[p]["kills"] += kills_r
        estadisticas[p]["asistencias"] += asist_r
        estadisticas[p]["puntos_totales"] += puntos_r

        #Clasificación por ronda: si el puntaje de esa ronda= 50 cuenta como pasar ronda
        if puntos_r >= 50:
            estadisticas[p]["rondas_pasadas"] += 1

#Guardar datos en diccionario y convertir a DataFrame
tabla = {
    "Jugador": [],
    "Kills": [],
    "Asistencias": [],
    "Puntos Totales": [],
    "Rondas Pasadas (por puntaje >=50/ ronda)": []
}

for p, v in estadisticas.items():
    tabla["Jugador"].append(p)
    tabla["Kills"].append(v["kills"])
    tabla["Asistencias"].append(v["asistencias"])
    tabla["Puntos Totales"].append(v["puntos_totales"])
    tabla["Rondas Pasadas (por puntaje >=50/ ronda)"].append(v["rondas_pasadas"])

df = pd.DataFrame(tabla)
df = df.sort_values(by="Puntos Totales", ascending=False).reset_index(drop=True)
df.index += 1
df["Clasificación"] = df.index

#Mostrar la tabla con Pandas
if has_display_tool:
    display_dataframe_to_user("Resultados del Torneo", df)
else:

    print("Resultados del Torneo:\n")
    print(df.to_string())

#Mostrar top 3 final
top3 = df.head(3).copy()
```
